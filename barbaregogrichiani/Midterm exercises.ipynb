# სავარჯიშო 4 - სტატისტიკა (5 ქულა)
### კუმულაციური ალბათობა (Cumulative distribution function)
დაითვალეთ შემდეგი კუმულაციური ალბათობა scipy.stats.norm.cdf() ფუნქციის გამოყენებით. დაითვალეთ მილიარდელების ჩამონათვლიდან რამდენ პროცენტია 50 წლის ზემოთ. 
norm.cdf(x, mean_val, std_dev_val) ფუნქციის პირველი პარამეტრია სასაზღვრო მნიშვნელობა, მეორე - საშუალო მნიშვნელობა, მესამე -სტდ. გადახრა.

მოახდინეთ შედეგების სიტყვიერი ინტერპრეტაცია.

იხილეთ დოკუმენტაციის ლინკი: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html

```python
from scipy.stats import norm
```

```python
df.age.describe()
```

```python
filt1 = df.age >50
```

```python
df.loc[filt1]
```

```python
df.loc[filt1].age.count()
```

```python
df.age.mean()
```

```python
df.age.std()
```

```python
x = 50
mean_val = df.age.mean()
std_dev_val = df.age.std()
```

```python
norm.cdf(x,mean_val,std_dev_val) # daaxl 9% aris 50wels qvemot 
```

```python
1-norm.cdf(x,mean_val,std_dev_val) #daaxl 91% aris 50 wels zemot
```
